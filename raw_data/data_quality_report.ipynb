### Dataset 1: Census Data

```python
census_data = pd.read_csv("census_data.csv")
census_data.head()
```

#### Completeness

```python
census_data.dtypes
```

The census data does not appear to have missing values. Since we converted all of the Census's default missing values to None/NaN in `fetch_census_data.py`, if there were missing values in `census_data.csv`, the dtypes for the column would indicate `object` or `float64`, not `str` or `int64`.

While this data does have complete data for all of the Micro/Metro areas that have a CBSA, it is not complete in the sense that there are significant portions of the United States that are not a part of a CBSA. Rural and sparsely populated areas will be excluded from this analysis simply because they do not a high enough population to be a part of a CBSA.

#### Accuracy

```python
census_data["name"].values.unique
```

This dataset does not appear to suffer from syntactic accuracy issues. The dataset has consistent capitalization and formatting in the `names` column.

Semantic accuracy is unfortunately difficult to check. For example, it would be challenging to check the total population of a specific CBSA without querying the census itself. The same can be said with most of the other columns in this dataset. There is not another authoritative source we can check to verify that the data in this dataset is correct. The population also is expected to change over time and may be suspect to drastic fluctuations, so checking this data against the older 2023 dataset for accuracy may not be a great way to check for semantic accuracy.

#### Consistency

One rule that this data must satisfy is that the column `population_25_plus` must be greater than or equal to the sum of `hs_diploma`, `ged`, `associates`, `bachelors`, `masters`, `professional`, and `doctorate`. We will check that in the code cell below.

```python
test = census_data[census_data["population_25_plus"] >= census_data["hs_diploma"] 
                   + census_data["ged"] + census_data["associates"]
                   + census_data["bachelors"] + census_data["masters"] 
                   + census_data["professional"] + census_data["doctorate"]]
assert(test.shape == census_data.shape)
print("Assertion Passed")
```

Another rule that this data must satisfy is that the column `total_population` must be greater than or equal to `population_25_plus`. We check this in the code cell below.

```python
test = census_data[census_data["total_population"] >= census_data["population_25_plus"]]
assert(test.shape == census_data.shape)
print("Assertion Passed")
```

Both rules are satisfied in this dataset.

#### Timeliness

This specific dataset will not change over time. We are using the 2024 American Community Survey data from the Census, which once it is published will not change. The values that we see now should be the same even into the future.

This dataset should be appropriate to use. This data is from the most recent published American Community Survey as of April 2026. It should remain appropriate to use for about the next year or until the next American Community Survey is published.

As we can see from the analysis above, we are missing values in some columns, specifically in the `CBSA` and the `NMCBSA` columns (which are the columns that we care the most about because we have to join on these columns). The notation "N" likely indicates the missing data.

The "N" likely means that the public school in question is probably not in a CBSA. We verify this using an example below:

```python
test = public_school_data[public_school_data["CBSA"] == "N"]
test.head()
```

Andalusia, AL, is located in Covington County. Covington County is not in a CBSA (according to [this map](https://www2.census.gov/geo/maps/metroarea/stcbsa_pg/Feb2013/cbsa2013_AL.pdf) of CBSAs in Alabama).

Similarly, Livingston, AL, is located in Sumter County. Sumter County is not in a CBSA according to the same map.

As a result, the missing values in the `CBSA` columns are highly likely due to the fact that these public schools are located in a region that is not encompassed by a CBSA. As a result, this missing data is likely intentional and not retrievable.

Ideally we should be more thorough and check every single entry, but doing so would require much more data (for example, every single city/county in every state) and is likely not feasible given the timeline of the project. This will be considered for future work.

#### Accuracy

```python
public_school_data["CITY"].value_counts()
```

```python
public_school_data["STREET"].value_counts()
```

This dataset suffers from both syntactic and semantic errors.

We can see that there is inconsistent capitalization and formatting in both the `CITY` and `STREET` columns. Sometimes, city names are capitalized, but other times they are not. "Road" is sometimes written "RD" or "Rd." depending on the entry. These contribute to syntactic errors.

The `STREET` column also suffers from semantic accuracy. For example, 13 public schools have their street address as simply "Main St" with no building number. This is a semantic error, as a school building should have a building number. We can see also that some schools also have multiple building numbers (for example, "9 1 and 12A").

While we can fix the syntactic errors, we do not see a way to fix the semantic errors in a timely manner, if at all. The address data of each school cannot be easily found without scraping websites of schools or asking the schools themselves.

#### Consistency

A rule that must be satisfied is that each CBSA code should match to exactly 1 NMCBSA. We check that below.

```python
cbsa_counts = public_school_data["CBSA"].value_counts()
nmcbsa_counts = public_school_data["NMCBSA"].value_counts()

for ((k1, _), (k2, _)) in zip(cbsa_counts.items(), nmcbsa_counts.items()):
    if k1 != "N":
        count_cbsa = cbsa_counts[k1]
        count_nmcbsa = nmcbsa_counts[k2]
        assert(count_cbsa == count_nmcbsa)
    else:
        assert(cbsa_counts[k1] == nmcbsa_counts[k2])

print("Assertions Passed")
```

The counts for each CBSA and NMCBSA match. While we can also check the other columns (namely `CSA` and `NMCSA`), we will remove those columns when we process the data in the data cleaning stage.

#### Timeliness

This particular dataset will not change over time. We are using the 2024-25 school location data from the Education Department, which should not change after the data is published. The data we see now should be the same into the future.

This data is from the 2024-25 school year, the most recent data available as of April 2026. This data should be suitable to use until the 2025-26 school year data is published by the Education Department.
